# Results Experiment 1

```python
import optuna
import joblib
import plotly.io as pio
pio.renderers.default = 'png'
study1 = joblib.load('study_exp_1.pkl', mmap_mode=None)
```

```python
fig = optuna.visualization.plot_contour(study1)
fig.show()
fig.write_image("exp1a.pdf", scale=1)
```

```python
fig = optuna.visualization.plot_param_importances(study1)
fig.show()
fig.write_image("exp1b.pdf", scale=1)
```

```python
fig = optuna.visualization.plot_edf(study1)
fig.show()
fig.write_image("exp1c.pdf", scale=1)
```

```python
fig = optuna.visualization.plot_parallel_coordinate(study1)
fig.show()
fig.write_image("exp1d.pdf", scale=1)
```
